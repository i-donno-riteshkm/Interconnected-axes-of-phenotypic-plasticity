## % better circuits for every score

```python
#for every score (row in the dataframe) calculating % of circuits that are better than WILD_type

df_score = pd.DataFrame(index=corr_data.index, columns=['% better circuits'])
print(df_score.shape)
df_score
```

```python
#for every score (row in the dataframe) calculating % of circuits that are better than WILD_type
for i in df_score.index:
    wild_type_score = float(corr_data.loc[i, 'a_new_network'])
    number_better_circuits = 0
    for j in corr_data.columns:
        if j == 'wild_type' or j == 'Network_4_new_edges_1':
            continue
        column_score = float(corr_data.loc[i, j])
        if (column_score > 0 and wild_type_score > 0) and (column_score > wild_type_score):
            number_better_circuits += 1
        elif (column_score < 0 and wild_type_score < 0) and (column_score < wild_type_score):
            number_better_circuits += 1
    df_score.loc[i, '% better circuits'] = number_better_circuits / 5
```

## No of better scores for every circuit

```python
# for every circuit calculate the number of scores that are better than wild type
df_network = pd.DataFrame(index=corr_data.columns, columns=['number of better scores'])
```

```python
for i in df_network.index:
    number_better_scores = 0
    for j in corr_data.index:
        score_wt = float(corr_data.loc[j, 'a_new_network'])
        score_col = float(corr_data.loc[j, i])
        if (score_col > 0 and score_wt > 0) and (score_col >= score_wt):
            number_better_scores += 1
        elif (score_col < 0 and score_wt < 0) and (score_col <= score_wt):
            number_better_scores += 1
    df_network.loc[i, 'number of better scores'] = number_better_scores

# sort the dataframe by the number of better scores
df_network = df_network.sort_values(by='number of better scores', ascending=False)
df_network.head(20)
```

```python
df_percentile = pd.DataFrame(index=corr_data.index, columns=['percentile'])
```

```python
from scipy.stats import percentileofscore

# find the %ile in which the wild type score is located for each score
for i in corr_data.index:
    wild_type_score = float(corr_data.loc[i, 'a_new_network'])
    # Calculate the percentile rank of the wild type score
    percentile_rank = percentileofscore(corr_data.loc[i, :], wild_type_score)
    if wild_type_score < 0:
        percentile_rank = 100 - percentile_rank
    print(f"{i}: Wild Type Score = {wild_type_score}, Percentile Rank = {percentile_rank:.2f}%")
    #fill the dataframe with the percentile rank
    df_percentile.loc[i, 'percentile'] = percentile_rank
df_percentile
```

```python
df_WT_percentile = pd.DataFrame(index = ["EM", "Glycolysis", "stemness", "Oxphos", "resistance", "Luminal-Basal"], 
                                columns=["EM", "Glycolysis", "stemness", "Oxphos", "resistance", "Luminal-Basal"])
# fill the dataframe with the percentile rank of the wild type score for each score
for i in df_percentile.index:
    all = i.split('_vs_')
    p = all[0]
    q = all[1]
    df_WT_percentile.loc[p, q] = df_percentile.loc[i, 'percentile']
    df_WT_percentile.loc[q, p] = df_percentile.loc[i, 'percentile']
```

```python
df_WT_percentile = df_WT_percentile.replace(np.nan, 100)
df_WT_percentile
```

```python
sns.heatmap(df_WT_percentile, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5, cbar_kws={"shrink": .8},
            vmin= 50, vmax=100)
plt.title('Wild Type Percentile Ranks for Each Score', fontsize=16)
```
